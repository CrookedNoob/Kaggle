# Missing Value Analysis

## Numeric Features

```python
total= numeric_features.isnull().sum().sort_values(ascending=False)
percent= (numeric_features.isnull().sum()/numeric_features.isnull().count()).sort_values(ascending=False)
missing_data= pd.concat([total, percent], axis=1, join='outer', keys=['Total NaN', '% NaN'])
missing_data.index.name='Numeric Feature'
missing_data.head(10)
```

```python
missing_values=numeric_features.isnull().sum(axis=0).reset_index()
missing_values.columns=['Features', 'NaN_Count']
missing_values= missing_values.loc[missing_values['NaN_Count']>0]
missing_values= missing_values.sort_values(by='NaN_Count')
missing_values
```

```python
ind=np.arange(missing_values.shape[0])
width=0.1
fig, ax= plt.subplots(figsize=(12, 5))
rects= ax.barh(ind, missing_values.NaN_Count.values, color='purple')
ax.set_yticks(ind)
ax.set_yticklabels(missing_values.Features.values)
ax.set_xlabel('NaN Counts')
ax.set_ylabel('Features')
plt.show()
```

```python
total= categorical_features.isnull().sum().sort_values(ascending=False)
percent= (categorical_features.isnull().sum()/categorical_features.isnull().count()).sort_values(ascending=False)
missing_data= pd.concat([total, percent], axis=1, join='outer', keys=['Total NaN', '% NaN'])
missing_data.columns=['Features', 'Nan_Count']
missing_data.index.name='Categorical Features'
missing_data.head(16)
```

```python
missing_values=categorical_features.isnull().sum(axis=0).reset_index()
missing_values.columns=['Features', 'NaN_Count']
missing_values= missing_values.loc[missing_values['NaN_Count']>0]
missing_values= missing_values.sort_values(by='NaN_Count')
missing_values
```

```python
ind= np.arange(missing_values.shape[0])
width=0.1
fig, ax= plt.subplots(figsize=(12,5))
rects= ax.barh(ind, missing_values.NaN_Count.values, color='olive')
ax.set_yticks(ind)
ax.set_yticklabels(missing_values.Features.values)
plt.xlabel('No. of Missing Values')
plt.ylabel('Features')
plt.show()
```
